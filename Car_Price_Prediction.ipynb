```python
cat = []
num = []

for i in df.columns:
    if df[i].dtype == 'O':
        cat.append(i)
    else:
        num.append(i)
```

```python
cat
```

```python
num
```

```python
for i in cat:
    vc = df[i].value_counts()
    plt.bar(vc.index, height = vc , color = 'c')
    plt.xlabel(i)
    plt.ylabel('Count')
    plt.show()
```

```python
# Numerical

for i in num:
    fig, ax = plt.subplots(1,2)
    ax[0].hist(df[i],color = 'c')
    ax[1].boxplot(df[i])
    plt.suptitle(i)
    plt.show()
```

```python
# iqr winsorization --
# replace lower outliers with lower fence
# replace upper outliers with upper fence


for i in num:
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    iqr = q3 - q1
    lf = q1 - 1.5*iqr
    uf = q3 + 1.5*iqr
    df.loc[df[i]<lf,i] = lf
    df.loc[df[i]>uf,i] = uf
```
